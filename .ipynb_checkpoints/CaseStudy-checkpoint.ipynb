## 單字詞詞向量實驗

### 目的

合併**多個多字詞向量**產生**一個單字詞向量**是否可行？

### Procedure: 

1. 找出所有「含有某個字」的詞彙
1. 用這些詞彙的詞向量做 clustering (by cosine similarity)
1. cluster 之後，最理想的情況是可以直接看出各個 cluster 的意義。

    在此情況下，將一個 cluster 中的所有詞向量合併成一個「平均向量」，檢視此「平均向量」是否真的與同義詞有高的 similarity。

## Function Definitions

```python
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
from numpy import dot
from numpy.linalg import norm

def query(char):
    global words
    lst = []
    for i in words:
        if (char in i): lst.append(i)
    return(lst)


def plot_dendro(cluster_data, method='average', metric='cosine', figsize=[20,20]):
    plt.figure(figsize=(figsize[0], figsize[1]))
    dend = shc.dendrogram(shc.linkage(cluster_data, method='average', metric='cosine'))


def cluster(cluster_data, n_clusters=5):

    # Cluster algorithm
    cluster = AgglomerativeClustering(n_clusters=n_clusters, affinity='cosine', linkage='average')
    cluster.fit_predict(cluster_data)

    # Print cluster label with word
    clustered = pd.DataFrame(list(zip(list(cluster_data.index), cluster.labels_)),
                             columns=['word', 'cluster'])

    for i in range(n_clusters):
        idx = clustered['cluster'] == i
        print(clustered[idx], '\n\n')
    
    # Return cluster object
    return(cluster)


# Print all words in a single cluster
def print_cluster(cluster_idx, cluster_data, cluster_obj):

    clustered = pd.DataFrame(list(zip(list(cluster_data.index), cluster_obj.labels_)),
                             columns=['word', 'cluster'])

    idx = clustered['cluster'] == cluster_idx
    print(clustered[idx].to_string())


# Return Mean Vector for each cluster (stored in a nested list)
def cluster_embeddings(cluster_data, cluster_labels):
    lst = []
    for i in list(set(cluster_labels)):
        mean_vector = cluster_data[cluster_labels == i].mean()
        lst.append(list(mean_vector))

    return(lst)

# Cosine similarity
def cos_sim(lst1, lst2, msg=''):
    cossim = dot(lst1, lst2)/(norm(lst1)*norm(lst2))
    print(msg, cossim)
    return(cossim)
```

## Prepare Data

```python
import pickle
with open('pd_embedings.pkl', 'rb') as f:
    wordvecs = pickle.load(f)

words = list(wordvecs.head(0)) # must exist for `query()` to work
```

```python
wordvecs[query('葡')].head()
```

## Hierachical Clustering

```python
cluster_data = wordvecs[query('葡')].T
```

### Dendrogram

```python
plot_dendro(cluster_data, figsize=[20,8])
```

### Clustering

```python
cluster_obj = cluster(cluster_data, n_clusters=3)
```

## Mean Vector of Each Cluster

```python
cluster_vecs = cluster_embeddings(cluster_data, cluster_obj.labels_)

# Mean vector of cluster 1
cluster_vecs[1]
```

# Case Studies

## 1. 猴 vs. 猿

假設：這兩個字的某個 Sense (Mean vector of one cluster) 的應該會有**高相似度** (cosine similarity)

### 猴

```python
cluster_data = wordvecs[query('猴')].T
plot_dendro(cluster_data, figsize=[20,8])
cluster_obj = cluster(cluster_data, n_clusters=9)

# Compute Mean Vector
cluster_vecs_monkey = cluster_embeddings(cluster_data, cluster_obj.labels_)
#cluster_vecs_monkey[1]
```

### 猿

```python
cluster_data2 = wordvecs[query('猿')].T
plot_dendro(cluster_data2, figsize=[20,8])
cluster_obj2 = cluster(cluster_data2, n_clusters=5)

cluster_vecs_primate = cluster_embeddings(cluster_data2, cluster_obj2.labels_)
#cluster_vecs_primate[1]
```

## Cosine Similarity: 猴 vs. 猿

```python
孫悟空_vec = list(wordvecs['孫悟空'])

cos_sim(cluster_vecs_monkey[1], cluster_vecs_primate[1], '猴1 vs 猿1：')
cos_sim(cluster_vecs_monkey[0], cluster_vecs_primate[1], '猴0 vs 猿1：')
cos_sim(cluster_vecs_monkey[1], cluster_vecs_primate[0], '猴1 vs 猿0：')
cos_sim(cluster_vecs_monkey[0], cluster_vecs_primate[0], '猴0 vs 猿0：')
print()
cos_sim(孫悟空_vec, cluster_vecs_monkey[1], '孫悟空 vs 猴1：')
cos_sim(孫悟空_vec, cluster_vecs_monkey[0], '孫悟空 vs 猴0：')
cos_sim(孫悟空_vec, cluster_vecs_primate[1], '孫悟空 vs 猿1：')
cos_sim(孫悟空_vec, cluster_vecs_primate[0], '孫悟空 vs 猿0：')
```

### Cluster 意義

- 猿0: 人名 or 地名 (e.g. 猿飛, 猿樂町)
- 猿1: 動物 (猿類)
- 猿2: ？
- 猿4: 棒球隊(桃猿)

<br>

- 猴0: 地名
- 猴1: 動物 (猴類)
- 猴8: 猴 + 人類文化

## 2. 抽象詞彙

### 2.1 好

預期：(至少) 看出破音字 (3 vs. 4 聲)

```python
cluster_data = wordvecs[query('好')].T
plot_dendro(cluster_data, figsize=[20,8])
cluster_obj = cluster(cluster_data, n_clusters=30)

# Compute Mean Vector
cluster_vecs = cluster_embeddings(cluster_data, cluster_obj.labels_)
```

```python
怪癖_vec = list(wordvecs['怪癖'])
興趣_vec = list(wordvecs['興趣'])

cos_sim(怪癖_vec, cluster_vecs[7], '好7 vs 怪癖：')
cos_sim(怪癖_vec, cluster_vecs[17], '好17 vs 怪癖：')
print()
cos_sim(興趣_vec, cluster_vecs[7], '好7 vs 興趣：')
cos_sim(興趣_vec, cluster_vecs[17], '好17 vs 興趣：')
```

#### 結果

- cluster 0: 數量龐大、多種意義
- cluster 7: 對應到 `CWN ㄏㄠˋ：喜愛並養成習慣`
- cluster 4,8,10: 大致對應到 `CWN 「好1」對特定對象的評價是正面的`
- cluster 12: 大致對應到 `CWN 「好3」親密`
- cluster 17: 對應到 `CWN ㄏㄠˋ：容易，常用於疾病`
- 其它: 新詞、專有名詞

### 2.2 飛

```python
cluster_data = wordvecs[query('飛')].T
plot_dendro(cluster_data, figsize=[20,8])
cluster_obj = cluster(cluster_data, n_clusters=35)

# Compute Mean Vector
cluster_vecs = cluster_embeddings(cluster_data, cluster_obj.labels_)
```

#### 結果

- cluster 1: 地名 (飛騨、飛驒國、郡飛島村)、特殊區域 (飛行區、飛機場、關西飛鳳園農場)
- cluster 2: 與「在空中運動」相關，如 飛躍、飛翔、飛機、飛彈、飛往、飛行員。參雜一點人名。
- cluster 5: 飛X，飛魚、飛沫、緋聞、飛鳶、飛蟲
- 其它：專有名詞

## 小結

- **抽象的詞彙或意義**，例如「好」、「飛」，很難 cluster 對應到 CWN 的分法。
- 名詞 (猿、猴) 較容易分出容易詮釋的類別。「猿」又比「猴」更容易 (因為使用環境較窄？)
