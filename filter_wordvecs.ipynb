## Load bin Modal from FastText

```python
from gensim.test.utils import datapath
from gensim.models._fasttext_bin import load

with open('fasttext/cc.zh.300.bin', 'rb') as fin:
    model = load(fin)
```

```python
print(model.nwords, model.vectors_ngrams.shape, sep='\t\t')
model.raw_vocab
```

```python
wordvec_freq = sorted(model.raw_vocab.items(), key=lambda tupl: tupl[1])
print(wordvec_freq[:20], wordvec_freq[-20:-1], sep='\n\n')
```

```python
# Save wordvec_freq to `word_fq_lst.pkl`
import pickle
with open('word_fq_lst.pkl', 'wb') as f:
    pickle.dump(wordvec_freq, f, protocol=4)
```

# 1. Get words that only consist of traditional Chinese characters

```python
# import word vec freq list
import pickle
with open('word_fq_lst.pkl', 'rb') as f:
    wordvec_by_freq = pickle.load(f)

# import Chinese Traditional Characters data
with open('as_chFreq.pickle', 'rb') as f:
    as_chFreq = pickle.load(f)
```

```python
print(list(as_chFreq)[:40])
trad_char_set = set(list(as_chFreq))
```

```python
# All words from FastText .bin
print(wordvec_by_freq[:20], wordvec_by_freq[-20:-1], sep='\n\n')
```

### Filter traditional Chinese words

```python
from itertools import compress

trad_ch_idx = [True]*len(wordvec_by_freq)
# loop over each word
for i in range(len(wordvec_by_freq)):
    word = wordvec_by_freq[i][0]
    
    # loop over each char j in word
    for char in word:
        if char not in trad_char_set:
            trad_ch_idx[i] = False
            break

# Traditional Chinese words
trad_wordvec = list(compress(data=wordvec_by_freq, selectors=trad_ch_idx))
trad_wordvec = sorted(trad_wordvec, key=lambda x: x[1], reverse=True)

print(len(trad_wordvec))
print()
print(trad_wordvec[:20], trad_wordvec[-20:-1], sep='\n\n')
```

## 2. Further Filtering by Freqency

```python
def filter_freq(data, freq):
    # Create an idx list for subsetting
    lt_freq_idx = [True]*len(data)
    for i in range(len(data)):
        if data[i][1] < freq:
            lt_freq_idx[i] = False
    
    return(list(compress(data, selectors=lt_freq_idx)))
```

```python
filtered_wordvec = filter_freq(trad_wordvec, freq = 1000)
print(len(filtered_wordvec), filtered_wordvec[-50:-1], sep='\n\n')
```

## 3. Filter WordVecs in `pd_embedings.pkl`

```python
with open('pd_embedings.pkl', 'rb') as f:
    wordvecs = pickle.load(f)
```

```python
wordvec_terms = [word[0] for word in filtered_wordvec]
filtered_wordvec_pd = wordvecs[wordvec_terms]
print(filtered_wordvec_pd.shape)
filtered_wordvec_pd.head()
```

```python
with open('pd_embedings_trad1000.pkl', 'wb') as f:
    pickle.dump(filtered_wordvec_pd, f, protocol=4)
```
